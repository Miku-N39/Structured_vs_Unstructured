```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_csv('runtimes_chiaraDATA.csv')
df
```

```python
seq_n4 = df['Seq_n4']
seq_4n = df['Seq_4n']
hil_n4 = df['Hil_n4']
hil_4n = df['Hil_4n']
ran_n4 = df['Ran_n4']
ran_4n = df['Ran_4n']
```

```python
# Mann-Whitney U Test
mann_whitney_test = stats.mannwhitneyu(seq_n4, seq_4n, alternative='two-sided')
print("Mann-Whitney U Test:", mann_whitney_test)
```

```python
# Mann-Whitney U Test
mann_whitney_test = stats.mannwhitneyu(hil_n4, hil_4n, alternative='two-sided')
print("Mann-Whitney U Test:", mann_whitney_test)
```

```python
# Mann-Whitney U Test
mann_whitney_test = stats.mannwhitneyu(ran_n4, ran_4n, alternative='two-sided')
print("Mann-Whitney U Test:", mann_whitney_test)
```

```python
label = ['nPoint, 4', '4, nPoint']

# Plotting the box-and-whisker plot
plt.figure(figsize=(8, 6))
plt.boxplot([seq_n4, seq_4n], labels=label)
plt.title(f'Comparison of lookup table process - Sequential', fontsize=16)
plt.ylabel('Runtime(sec)', fontsize=12)
plt.savefig('Seqential')
plt.show()
```

```python
label = ['nPoint, 4', '4, nPoint']

# Plotting the box-and-whisker plot
plt.figure(figsize=(8, 6))
plt.boxplot([hil_n4, hil_4n], labels=label)
plt.title(f'Comparison of lookup table process - Hilbert',  fontsize=16)
plt.ylabel('Runtime(sec)',  fontsize=12)
plt.savefig('Hilbert')
plt.show()
```

```python
label = ['nPoint, 4', '4, nPoint']

# Plotting the box-and-whisker plot
plt.figure(figsize=(8, 6))
plt.boxplot([ran_n4, ran_4n], labels=label)
plt.title(f'Comparison of lookup table process - Random',  fontsize=16)
plt.ylabel('Runtime(sec)',  fontsize=12)
plt.savefig('Random')
plt.show()
```
